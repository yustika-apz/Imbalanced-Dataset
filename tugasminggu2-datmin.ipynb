```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

df = pd.read_csv('../input/covid19-in-india/covid_19_india.csv')
df
```

```python
Tripura = df[df['State/UnionTerritory'] == 'Tripura']
Tripura
```

```python
kasus = Tripura[Tripura['Confirmed']> 0]
kasus
```

```python
plt.plot(kasus['Date'] , kasus['Confirmed'])
plt.title('Kasus Covid di Tritura India 2020-2021')
plt.xlabel('Date')
plt.ylabel('Jumlah Pasien Terkonfirmasi Terdampak Covid')
plt.show()
```

```python
kasusbaru = kasus[kasus.Date >='2021-01-01']
kasusbaru
```

```python
plt.plot(kasusbaru.Date , kasusbaru.Confirmed)
plt.plot(kasusbaru.Date , kasusbaru.Cured)
plt.plot(kasusbaru.Date , kasusbaru.Deaths)
plt.legend(['Jumlah Pasien Terkonfirmasi', 'Jumlah Pasien yang Sembuh', 'Jumlah Pasien yang Meninggal'])
plt.title('Jumlah Pasien Covid di Tritura India Tahun 2021')
plt.xlabel('Date')
plt.show()
```
